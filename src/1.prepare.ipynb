```python
def isNan(line):
    if line.find('?') > 0:
        return 'Indefinite'
    elif len(line.strip()) == 0:
        return 'Indefinite'
    else:
        return line.strip().upper()
```

```python
df['sex_'] = df['sex'].apply(lambda x: isNan(x))
df['race_'] = df['race'].apply(lambda x: isNan(x))
df['native_'] = df['native-country'].apply(lambda x: isNan(x))
```

```python
def income(line):
    if (line.strip()) == '<=50K':
        return 0
    else:
        return 1
```

```python
cols_num = []
for c in df_prepare.columns:
    t = df_prepare[c].dtype
    if t != 'object':
        cols_num.append(c)
    print(c, '-', t)
```

```python
cols_num
```

```python
cols = []
for c in df_prep_final.columns:
    c = c.replace('(','_')
    c = c.replace(')','_')
    c = c.replace('&','_')
    c = c.replace('_','-')    
    cols.append(c.lower())
```
